```python
import torch
import torch.nn as nn
import torch.optim as optim

class QMFEmbeddingLayer(nn.Module):
    def __init__(self, input_dim, embed_dim, lambda_reg):
        super(QMFEmbeddingLayer, self).__init__()
        self.embed_dim = embed_dim
        self.lambda_reg = lambda_reg
        
        # Calculate the dimension for QMF transformation
        self.qmf_dim = (embed_dim + 1) + (embed_dim * (embed_dim + 1)) // 2
        
        # Initial embedding layer
        self.embeddings = nn.Embedding(input_dim, embed_dim)
        
        # QMF parameters
        self.R = nn.Parameter(torch.randn(embed_dim, self.qmf_dim))
        self.J = torch.eye(self.qmf_dim)  # Example J matrix
        
    def forward(self, x):
        # Get initial embeddings
        embeds = self.embeddings(x)
        
        # Apply quadratic transformation
        T_embeds = self.quadratic_transform(embeds)
        
        # Compute QMF reconstruction
        X_hat = torch.matmul(T_embeds, self.R.t())
        
        # Compute QMF loss
        qmf_loss = torch.norm(X_hat - embeds, p='fro') ** 2 + self.lambda_reg * torch.norm(self.R @ self.J, p='fro') ** 2
        
        return embeds, qmf_loss
    
    def quadratic_transform(self, embeds):
        batch_size, embed_dim = embeds.size()
        
        # Include linear terms
        quadratic_terms = [embeds]
        
        # Include quadratic terms
        for i in range(embed_dim):
            for j in range(i, embed_dim):
                quadratic_terms.append((embeds[:, i] * embeds[:, j]).unsqueeze(1))
        
        # Concatenate the bias term, linear terms, and quadratic terms
        T_embeds = torch.cat([torch.ones(batch_size, 1, device=embeds.device)] + quadratic_terms, dim=1)
        
        return T_embeds

# Example usage
input_dim = 10000  # Vocabulary size
embed_dim = 300    # Embedding dimension
lambda_reg = 0.1   # Regularization parameter

qmf_layer = QMFEmbeddingLayer(input_dim, embed_dim, lambda_reg)

# Example input batch
x = torch.randint(0, input_dim, (32,))

# Forward pass
embeds, qmf_loss = qmf_layer(x)
print(f"Embeddings shape: {embeds.shape}")
print(f"QMF loss: {qmf_loss.item()}")
```
